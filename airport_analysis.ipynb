## Airport Traffic Analysis

```python
#Which origin airports handle most outgoing traffic?
origin_traffic = df.groupby("ORIGIN_CITY_NAME")["PASSENGERS"].sum().sort_values(ascending=False)

top_origin_airports = origin_traffic.head(10).reset_index()

top_origin_airports.columns = ["Origin_Airport", "Outgoing_Passengers"]

top_origin_airports
```

```python
plt.figure(figsize=(12,6))

top_origin_airports.plot(kind="bar", x="Origin_Airport", y="Outgoing_Passengers",legend=False)

plt.title("Top 10 Origin Airports by Outgoing Passenger Traffic", fontsize=14)
plt.xlabel("Origin Airport", fontsize=12)
plt.ylabel("Outgoing Passengers", fontsize=12)
plt.xticks(rotation=45, ha="right")

plt.show()
```

### Insight:
Atlanta, GA recorded the highest outgoing passenger traffic, followed by Denver and Chicago. This indicates that these cities act as major aviation departure hubs with strong passenger connectivity and travel demand.

```python
#Which destination airports receive the most incoming traffic?
dest_traffic = df.groupby("DEST_CITY_NAME")["PASSENGERS"].sum().sort_values(ascending = False)

top_dest_airports = dest_traffic.head(10).reset_index()

top_dest_airports.columns = ["Destination_Airport", "Incoming_Passengers"]

top_dest_airports
```

```python
plt.figure(figsize=(12,6))

top_dest_airports.plot( kind="bar", x="Destination_Airport",y="Incoming_Passengers",legend=False)

plt.title("Top 10 dest Airports by incoming Passenger Traffic", fontsize=16)
plt.xlabel("Destination Airport", fontsize=12)
plt.ylabel("Incoming Passengers", fontsize=12)
plt.xticks(rotation=45, ha="right")

plt.show()
```

### Insight:
Atlanta, GA received the highest incoming passenger traffic among all destination airports, followed by Chicago and Denver. This suggests that these cities serve as major arrival hubs with strong air connectivity and high passenger demand.

The analysis highlights the importance of metropolitan airports in handling large volumes of inbound passenger movement across the aviation network.

```python
# top-performing Airport

total_airport_traffic = origin_traffic.add(dest_traffic, fill_value=0)

#sorting the total airport traffic from highest to lowest
total_airport_traffic = total_airport_traffic.sort_values(ascending=False)

print("Top 10 airport as per the total passenger movement : ")
top_10_airports = total_airport_traffic.head(10).apply(lambda x: f"{x:,.0f}").reset_index()

top_10_airports.columns = ["Airport_City", "Total_Passenger_Movement"]

top_10_airports
```

```python
top_10_airports = total_airport_traffic.head(10)

plt.figure(figsize=(12,6))
top_10_airports.plot(kind="bar")
plt.title("Top 10 Airports by Passenger Movement")
plt.xlabel("Airport City")
plt.ylabel("Total Passengers")
plt.xticks(rotation=45)

plt.show()
```

### Insight:
Atlanta, GA emerged as the top-performing airport based on total passenger movement, followed by Denver and Chicago. These airports handled the highest combined incoming and outgoing passenger traffic, highlighting their role as major aviation hubs within the network.

The analysis indicates that airports located in large metropolitan cities experience significantly higher passenger movement due to strong connectivity, business travel, tourism activity, and extensive airline operations.

## Airline Analysis

```python
#Top Airlines by Passenger Count

top_airlines = df.groupby("UNIQUE_CARRIER_NAME")["PASSENGERS"].sum()

#sorting the values from high to low
top_airlines = top_airlines.sort_values(ascending=False)

#top 10 airlines
top_airlines.head(10).apply(lambda x: f"{x:,.0f}")
```

```python
top_10_airlines = top_airlines.head(10)

plt.figure(figsize=(12,6))

top_10_airlines.plot(kind="bar")

plt.title("Top 10 Airlines by Passenger Count", fontsize=16)
plt.xlabel("Airline", fontsize=12)
plt.ylabel("Total Passengers", fontsize=12)
plt.xticks(rotation=45)

plt.show()
```

### Insight:
Southwest Airlines Co. carried the highest number of passengers among all airlines in the dataset, followed by Delta Air Lines and American Airlines. This indicates that these airlines dominate passenger transportation and maintain extensive operational networks across major routes.

The analysis also shows that a few large airlines account for a significant share of total passenger traffic, highlighting strong market concentration within the aviation industry.

```python
#Which airlines typically fly longer routes?
avg_distance_airlines = df.groupby("UNIQUE_CARRIER_NAME")["DISTANCE"].mean().sort_values(ascending=False)

#top 10 

top_distance_airlines = avg_distance_airlines.head(10).reset_index()
top_distance_airlines.columns = ["Airline", "Average_Distance"]
top_distance_airlines
```

```python
top_distance_airlines.plot(kind="bar", x="Airline", y="Average_Distance",legend=False
)

plt.title("Top Airlines by Average Flight Distance",fontsize=16)
plt.xlabel("Airline", fontsize=12)
plt.ylabel("Average Distance", fontsize=12)
plt.xticks(rotation=45, ha="right")

plt.show()
```

### Insight:
Asia Pacific and VistaJet Limited recorded the highest average flight distances among all airlines, indicating a stronger focus on long-haul operations. Airlines such as Hawaiian Airlines and Kalitta Air also operated significantly longer routes compared to other carriers in the dataset.

The analysis suggests that these airlines primarily serve international, cargo, or long-distance travel corridors requiring extended flight operations.

## Route Analysis

```python
#creating a route column

df["ROUTE"] = df["ORIGIN_CITY_NAME"] + " - " + df["DEST_CITY_NAME"]
#Which airport routes carry the maximum passengers?

busiest_routes = df.groupby("ROUTE")["PASSENGERS"].sum().sort_values(ascending=False).head(10).reset_index()
busiest_routes.columns = ["Routes", "Total Passenger"]
busiest_routes
```

```python
busiest_routes.plot(kind="barh",x="Routes", y="Total Passenger",legend=False)

plt.title("Top 10 Busiest Flight Routes")
plt.xlabel("Route")
plt.ylabel("Passengers")
plt.xticks(rotation=75)

plt.show()
```

### Insight:
Routes connecting New York with Miami, Orlando, Chicago, and Atlanta recorded the highest passenger traffic in the dataset. This highlights New York as a major aviation hub with extremely strong intercity connectivity and travel demand.

The analysis also indicates that routes between large metropolitan and tourism-focused cities experience significantly higher passenger movement due to business travel, tourism activity, and dense flight operations.

```python
#Which city routes are the longest by distance?
longest_routes = df.groupby("ROUTE")["DISTANCE"].mean().sort_values(ascending=False)
top_longest_routes = longest_routes.head(10).reset_index()
top_longest_routes.columns = ["Routes", "Average Distance"]
top_longest_routes
```

```python
longest_routes.head(10).sort_values().plot(kind="barh")

plt.title("Top 10 Longest Flight Routes")

plt.xlabel("Average Distance")

plt.ylabel("Route")

plt.show()
```

### Insight:
Routes connecting Hawaii with mainland US cities such as Boston, New York, and Newark recorded the longest average flight distances in the dataset. The Honolulu–Boston route emerged as the longest corridor, covering more than 5,000 miles on average.

The analysis highlights the operational importance of long-haul domestic routes connecting geographically distant regions, particularly Hawaii, with major metropolitan cities in the continental United States.

## Geographic Analysis

```python
#Which states generate highest passenger traffic?

origin_state_traffic = df.groupby("ORIGIN_STATE_NM")["PASSENGERS"].sum()
dest_state_traffic = df.groupby("DEST_STATE_NM")["PASSENGERS"].sum()
total_state_traffic = origin_state_traffic.add(dest_state_traffic, fill_value=0).sort_values(ascending=False).head(10).reset_index()
total_state_traffic.columns = ["State", "Passenger Traffic"]
total_state_traffic
```

```python
total_state_traffic.head(10).plot(kind="barh",x="State", y="Passenger Traffic" )

plt.title("Top States by Total Passenger Traffic")

plt.xlabel("Passengers")

plt.ylabel("State")

plt.show()
```

### Insight:
Florida generated the highest passenger traffic among all states, followed closely by California and Texas. These states represent major aviation markets due to their large populations, strong tourism industries, extensive business activity, and high airport connectivity.

The analysis also shows that states containing major metropolitan cities and tourism destinations contribute significantly to overall passenger movement within the aviation network.

```python
#Which regions are busiest?

region_traffic = df.groupby("REGION")["PASSENGERS"].sum().sort_values(ascending=False)
region_traffic.plot(kind="pie", autopct="%1.1f%%")

plt.title("Passenger Traffic Distribution by Region")

plt.ylabel("")

plt.show()
```

### Insight:
The dataset was overwhelmingly dominated by domestic aviation traffic, with the Domestic (D) region accounting for nearly all passenger movement. This indicates that the dataset primarily represents domestic flight operations rather than international air travel.

The analysis suggests that domestic routes form the core of passenger transportation activity within the dataset.
